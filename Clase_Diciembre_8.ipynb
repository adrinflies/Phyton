# Operaciones Matematicas

```
# 4to nivel de importancia
# Suma Ordinaria
print("-------------------")
print("Suma Ordinaria")
print(3 + 4) # un 7

# Declaración de variables
variable1 = 20
variable2 = 40
print(variable1 + variable2)

# Resta Ordinaria
print("-------------------")
print("Resta Ordinaria")
print(2 - 5)

# Declaración de variables
v1 = 10.4
v2 = 15
resultado = v2 - v1
print(resultado) # 4.6
```

```
# 3er Nivel de Importancia
# Multiplicación Ordinaria
print("Multiplicación Ordinaria")
M1, M2 = 2,3.14
print(M1 * M2) #6.25

# Divisón Ordinaria
print("-------------------")
print("División Ordinaria")
M1,M2,M3,M4,M5,M6 = 2,5,6,5,4,5
print(M1 / M2)
print(M3 / M4)
print(M5 / M6)

# Division de Piso
print("-------------------")
print("División de Piso")
print(M1 // M2)
print(M3 // M4)
print(M5 // M6)
```

```
# Potencias
print("Potencias")
print(2 ** 3) # es 8

# Declaración de Variables
P1,P2 = 7,2
print(P1 ** P2) # es 49
```

```
# Area de un circulo
print("Area de un Circulo")
pi = 3.14
r = 4
A = pi * (r ** 2)
print(A)
```
