## Funciones
En este curso hemos usado algunas funciones integradas en phyton, pero tambien podemos definir funciones, siendo estas un bloque de codigo reutilizable que podemos mandar a llamar varias veces en el codigo.

```python
# Descuento
def descuento(precio, desc = 0.15):
  precio = precio * (1 - desc)
  print("El precio con descuento es: ", precio)
descuento(750)
descuento(800, .25)
```

```python
def usuario(nombre,edad):
  return (nombre,edad)
persona = usuario("Adrián", 19)
print(type(persona))
print(persona)
```

```python
def operaciones(numero):
  doble = numero * 2
  triple = numero * 3
  cuadruple = numero * 4
  quintuple = numero * 5
  return doble, triple, cuadruple, quintuple
resultado = operaciones(4)
print(resultado)
```

#### Funciones que llaman a otras funciones

```python
def saludar(nombre):
  return f"Hola, {nombre}"

def despedirse(nombre):
  return f"Adios, {nombre}"

def mensaje_completo(nombre):
  saludo = saludar(nombre)
  despedida = despedirse(nombre)
  return saludo + "\n" + despedida
print(mensaje_completo("Emmanuel"))
```

## Funciones Lambda
Una funcion lambda es una función acortada para escribir una función de esta forma:

nombreFuncion = lambda parametros: acción

La diferencia entre las funciones lambda y las funciones normales es que suelen ser un poco mas limitadas.

```python
sumar = lambda x,y : x+y
print(sumar(3,10))
```

```python
descuentoLambda = lambda desc, precio: ((precio * (1-desc)), precio * desc)
nuevoPrecio = descuentoLambda(0.4,500)
print("El precio a pagar es", nuevoPrecio[0], ", el descuento que se hizo fue de: ", nuevoPrecio[1])
print(type(nuevoPrecio))
```

```python
# Calculadora -> suma, resta, producto divison y potencia
def suma(a,b):
  return a + b
def resta(a,b):
  return a - b
def producto(a,b):
  return a * b
def division(a,b):
  if b == 0:
    return "Error no se puede dividir entre 0"
  return a / b
def potencia(a,b):
  return a ** b

def menu():
  print("-----CALC (short for calculator)-----")
  print("1. Suma")
  print("2. Resta")
  print("3. Producto")
  print("4. Division")
  print("5. Potencia")
  print("6. Salir")

while True:
  menu()
  try:
    op = int(input("Seleccione una opción: "))
    if op < 1 or op > 6:
      print("Opción invalida")

  except ValueError:
    print("Ingrese un numero valido")
  if op == 6:
    print("SALIENDO....")
    break

  try:
    a = float(input("Ingrese el primer numero: "))
    b = float(input("Ingrese el segundo numero: "))

  except ValueError:
    print("Error: Ingrese un numero valido")
  if op == 1:
    print("Resultado:", suma(a,b))
  elif op == 2:
    print("Resultado:", resta(a,b))
  elif op == 3:
    print("Resultado:", producto(a,b))
  elif op == 4:
    print("Resultado:", division(a,b))
  elif op == 5:
    print("Resultado:", potencia(a,b))
  else:
    print("Opción Invalida")
```
