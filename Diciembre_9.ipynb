## If/Else

```python
# Notas de Materias
nota = float(input("Ingresa una calificación: "))

if nota >= 70: # si es mayor o igual a 70
  print("Acreditaste la materia")
else: # si es menor a 70
  print("Reprobaste")
```

```python
edad = int(input("Cual es tu edad: "))
if edad >= 18: # mayor a 18
  print("Eres mayor de edad")
  if edad >= 65: # mayor a 65
    print("Ademas ya eres adulto mayor")
  else: # menor a 65
    print("Eres un adulto, pero no mayor")
else: # menor a 18
  print("Eres menor de edad")
  if edad <= 12: # menor a 12
    print("Eres un niño")
  else: # mayor a 12
    print("Eres adolescente")
```

## Elif
Haremos un programa en el que preguntaremos la calificacion del 0 al 100 y queremos saber su equivalente en el sistema de evaluacion estadounidense:
- A+ = 100
- A = 90 - 99
- B = 80 - 89
- C = 70 - 79
- D = 60 - 69
- F = < 60

```python
cal = 85
if cal == 100:
  print("A+")
elif cal >= 90:
  print("A")
elif cal >= 80:
  print("B")
elif cal >= 70:
  print("C")
elif cal >= 60:
  print("D")
else: # como al ya no haber una condicion, el else cubre todo lo demas
  print("F")
```

## Operaciones Logicas con Condiciones

```python
# and, or, not
# Quieres un sandwich con todo
pan = True
queso = True
mayonesa = True
jamon = False
print(pan and queso and mayonesa and jamon)
if not (pan and queso and mayonesa and jamon):
  print("No te puede hacer el sandwich")
```

```python
cal = int(input("Cual es tu calificación: "))
if cal >= 0 and cal <= 100:
  if cal == 100:
    print("A+")
  elif cal >= 90:
    print("A")
  elif cal >= 80:
    print("B")
  elif cal >= 70:
    print("C")
  elif cal >= 60:
    print("D")
  else: # como al ya no haber una condicion, el else cubre todo lo demas
    print("F")
else:
  print("Error: La calificación debe de estar entre 0 y 100")
```

```python
# EL modulo (%)
# EL modulo es un operador matematico que devuelce el residuo de una división
# 10 % 3 da 1, por que 10 dividido entre 3 cabe 3 veces y sobra 1
# 7 % 2 da 1, por que el 7 no es divisible de forma exacta entre 2
# 2 % 2 -> 0

num = 9
if num % 3 == 0 or num % 5 == 0:
  print(f"El numero {num} es multiplo de 3 o de 5")

# saber si el numero es par o impar
num = 6
if not num % 2 == 0:
  print(f"El numero {num} es impar")
```

## Ciclo for
Los llamados ciclos for, nos ayudan a realizar acciones durante una cantidad determinada de veces

```python
# Imprimir los numeros del 0 al 4
for i in range(5): # toma 5 como el numero final pero inicia siempre en 0
  print(i)
```

```python
for num in range (2,11):
  print(num)
```

```python
# Tanbien existe el range(inicio, fin, paso)
for i in range(2,21,2):
  print(i)
```

```python
cadena = "Ingeniero Biomedico"
for letra_por_letra in cadena:
  print(letra_por_letra)
```

## While

```python
suma = 0
numeros = int(input("Ingresa el numero para sumar o 0 para salir: "))

while numeros != 0: # != es equivalente a "distinto a 0"
  suma += numeros
  numeros = int(input("Ingresa el numero para sumar o 0 para salir: "))

print(f"Ty suma es: {suma}")
```

```python
# usar el while como for
numero = 0
while numero <= 12:
  print(numero)
  numero += 1
```
