## Estructuras de Datos
Las estructuras de datos son formas de organizar informacipón para trabajar con ellas de manera eficiente.

- Listas -> ordenadas, modificadas, permiten duplicación
- Tuplas -> ordenadas, inmutables, permiten duplicados
- Sets -> no ordenados, sin duplicados
- Diccionarios -> pares clave:valor

Nota:
- Pilas (FILO -> First In, Last Out)
- Colas (FIFO -> First In, First Out)
- Listas enlazadas (o doblemente enlazadas)

## Listas
Las listas son condiciones ordenadas que podemos modificar y almacenan elementos de distintos tipos de datos.

```python
lista_de_compras = ["manazas", "pan", "leche", "huevos", "mantequilla", "queso", "jamon"]
print(lista_de_compras)
```

```python
print(lista_de_compras[4])
```

```python
# Agregar un elemento a la lista de compras
lista_de_compras.append("sopa")
lista_de_compras
```

```python
lista_de_compras.insert(1, "platano")
lista_de_compras
```

```python
print(lista_de_compras)
```

```python
# Para eliminar un elemento de una lista, existe el remove y el pop
lista_de_compras.remove("sopa")
lista_de_compras.remove("platano")
print(lista_de_compras)
```

```python
lista_de_compras.pop() # Sin nada dentro del parentesis borra el ultimo elento
lista_de_compras.pop(2) # o puedes agarrar la posición que quiera (iniciando desde 0)

# Nota: Te retorna el numero eliminado
```

```python
# Podemos recorrer el arrgelo por medio de un "for"
for elementos in lista_de_compras:
  print(elementos)
```

```python
# len() saca el tamaño de la lista
print(len(lista_de_compras))
for i in range(len(lista_de_compras)):
  print(f"{i + 1} - {lista_de_compras[i]}")
```

```python
promedios = [55.6,97.5,23.6,97.5,72.2,100]
print(promedios)
promedios.sort() # sort() ordena un set de valores de manera ascendente
print(promedios)

# Ordenar una lista de manera descendente
promedios.sort(reverse = True)
print(promedios)
```

```python
# Podemos ingresar al utlimo elemento de la lista
print(promedios[-1])
```

```python
# Encontrar el mayor y el menor de una lista
print(max(promedios))
print(min(promedios))
```

```python
# Declaración y usos de matrices
calificaciones = [[100,88,93],[99,56,72]]
i = 1
for cali_i in calificaciones:
  print(f"\nAlumno {i}")
  j = 1
  for cali_j in cali_i:
    print(f" Calificación {j}: {cali_j}")
    j += 1
  j = 0
  i += 1
```

## Sets
Los sets son colecciones SIN ORDEN y que NO permiten elementos duplicados

```python
nombre = {"Adrián", "Alex", "Santiago", "Emmanuel", "Emmanuel"}
print(nombre)
print(type(nombre))
```

```python
distintos_datos = {"Alex", 1, "Santiago", "Emmanuel", "Emmanuel", 4.8}
distintos_datos
```

```python
# Add
nombre.add("Gael")
nombre
```

```python
print(nombre[2])
```

```python
nombre[2] = "Hugo"
```

```python
# Union
matutino = {2111735,244458,111111,4568887}
vespertino = {2111735,2123443,3687467,4568887}

# Union pero con punto .union()
total_matriculas = matutino.union(vespertino)
print(total_matriculas)

# Union pero con |
total_matriculas = matutino | vespertino
print(total_matriculas)
```

```python
# Intersección
ca = {2111735,244458,111111,4568887}
al = {2111735,111111}

# Intersección con .intersection()
ca_al = ca.intersection(al)
print(ca_al)

# Intersección usando &
ca_al = ca & al
print(ca_al)
```

```python
# Diferencia
# Diferencia usando .difference
solo_calculo = ca.difference(al)
print(solo_calculo)

# Diferencia usando -
solo_calculo = ca - al
print(solo_calculo)
```

```python
nombre = {"Adrián", "Alex", "Santiago", "Emmanuel", "Emmanuel"}
# remove, pop
nombre.remove("Santiago") # Deber seleccionar el que tu quieras eliminar
print(nombre)
nombre.pop()
print(nombre)
```

```python
# Elimninar todo el SET
nombre.clear()
nombre
```

## Diccionarios
Son colecciones no ordenadas de pares clave:valor

```python
alumno = {
    #clave    valor
    "nombre": "Adrián Noriega Solís",
    "edad": 19,
    "carrera": "Ingenieria Biomedica",
    "semestre": 6,
    "promedio": 85,
    "calificaciones": [89,100, 72, 71, 86]
}
print(type(alumno))
alumno
```

```python
print("El alumno ", alumno["nombre"], " esta estudiando ", alumno["carrera"])
```

```python
# Agregar al diccionario
alumno["area preferida"] = "Mecanica"
alumno
```

```python
for valor in alumno:
  print(alumno[valor])
```

```python
for clave, valor in alumno.items(): # devuelve una vista de todos los pares clave:valor del diccionario
  print(clave, ": ", valor)
```

```python
# El metodo .clean() tambien usar aqui
# Tenemos el .get que nos ayuda a obtener el valor (value) de cierta clave (key)
print(alumno.get("nombre"))
```

```python
alumno_1 = alumno.items()
print(alumno_1)
print(type(alumno_1))
alumno_lista = list(alumno_1)
print(alumno_lista)
```

```python
# El metodo .values() sirve para obtener los values del diccionario
alumno_v = alumno.values()
print(type(alumno_v))
alumno_v
```

```python
if "carrera" in alumno:
  print("Existe carrera")
else:
  print("No existe carrera")

if("dinero") in alumno:
  print("Aluno tiene dinero")
else:
  print("Alumno no tiene dinero")
```

```python
# .copy() puedes copiar todo el diccionario en una sola variable
# Update -> Sirve para cambiar los values (valores) (.update())
alumno.update({"semestre":7})
alumno
```
